# 2. MAGIC imputation

Count data are first notmalized for library size and square-root transformed. We then aply MAGIC (Markov Affinity-based Graph Imputation of Cells), an algorithm for denoising and imputation of single cells applied to single-cell RNA sequencing data, as described in Van Dijk D et al. (2018), Recovering Gene Interactions from Single-Cell Data Using Data Diffusion, Cell. [https://www.cell.com/cell/abstract/S0092-8674(18)30724-4]

```python
import magic
data_libnorm, libsize = scprep.normalize.library_size_normalize(adata.X, return_library_size=True)
data_sqrt = scprep.transform.sqrt(data_libnorm)
data_magic = magic.MAGIC().fit_transform(data_sqrt)
```

```python
adata.layers["magic"]=data_magic
```

```python
gene_list = adata.var.index.values
dge = pd.DataFrame(data_magic,columns = gene_list)
pd.DataFrame(dge).to_csv("../../data/data03-scvi/ut_0-adjLogcounts_magic.csv")
```

# 3. scVI

```python
adata.layers["counts"] = adata.X.copy() # preserve counts
sc.pp.normalize_total(adata, target_sum=1e4)
sc.pp.log1p(adata)
adata.raw = adata # freeze the state in `.raw`
```

```python
adata
```

scVI is a hierarchical Bayesian model for single-cell RNA sequencing data with conditional distributions parametrized by neural networks. Working as a hybrid between a neural network and a bayesian network, scVI performs data harmonization. 

Below we train scVI's variational auto-encoder, compute and store the latent representation of the data

```python
import scvi
```

```python
scvi.data.setup_anndata(
    adata,
    layer="counts",
    batch_key='batch',
    continuous_covariate_keys=["log_counts"]
)
```

```python
model = scvi.model.SCVI(adata)
```

```python
model
```

```python
model.train()
```

```python
model
```

```python
# model.save("../../data/data03-scvi/my_model_ut_0-adjLogcounts/", save_anndata = True)
model = scvi.model.SCVI.load("../../data/data03-scvi/my_model_ut_0-adjLogcounts/", use_gpu=True)
```

```python
adata.layers["scvi_normalized"] = model.get_normalized_expression(
    library_size=10e4
)
```

```python
latent = model.get_latent_representation()
adata.obsm["X_scVI"] = latent
```

```python
sc.pp.neighbors(adata, use_rep="X_scVI")
sc.tl.umap(adata, min_dist=0.3)
```

```python
sc.pl.umap(
    adata,
    color=["orig.ident"],
    palette=['xkcd:dirty blue', 'xkcd:tangerine', 'xkcd:dark seafoam'],
    frameon=False,
)
```

```python
# neighbors were already computed using scVI
sc.tl.leiden(adata, key_added="leiden_scVI", resolution=0.4)
```

```python
sc.pl.umap(
    adata,
    color=["leiden_scVI", "log_counts"],
    frameon=False,
)
```

# 4. Explore expression of signature markers across clusters (Figure 2G)

```python
sc.tl.dendrogram(adata, groupby="leiden_scVI", use_rep="X_scVI")

markers_sig = ["HSP70", "HSP90",
              "HTA2", "HHF1",
               "TTR1", "TRX1", ]

sc.pl.dotplot(
    adata,
    markers_sig,
    groupby='leiden_scVI',
    dendrogram=False,
    color_map="Blues",
    categories_order = ["2", "0", "1"],
    swap_axes=True,
    use_raw=True,
    standard_scale="var",
    save="figure2g-ut_sig_markers.pdf"
)
```
